# 點歌機進階版

```python
from IPython.display import YouTubeVideo
```

```python
happy= YouTubeVideo('o2h6ly4mCm8')
sad= YouTubeVideo('ZwguVwoDsvs')
soso= YouTubeVideo('0ustDXbQH1I')
hard= YouTubeVideo('IEcb-FB1iLQ')
miss= YouTubeVideo('rhA2BwdKiBI')
peace= YouTubeVideo('eWscFcv8A5g')
lonely= YouTubeVideo('ELERFreNG5Q')
```

```python
msg="我只想讓你知道我都在"
print( "(≧▽≦)：今天想聽什麼歌？")
while(msg!="謝謝"):
    msg=input(">>")
    if("謝謝"in msg):
        print("(≧▽≦)：下次再來")
    elif("人生好難"in msg):
        display(hard)
    elif("有點想家"in msg):
        display(miss)
    elif("難過"in msg):
        display(sad)
    elif("開心"in msg):
        display(happy)
    elif("平靜"in msg):
        display(peace)
    elif("一個人"in msg):
        display(lonely)
    elif("還好"in msg):
        display(soso)
```
